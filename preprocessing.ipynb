#### Convert audio to speaker-specific transcript
Now, we need to convert this audio to a transcript that is speaker specific. We will use pyannote for this, following along with [this tutorial](https://ngwaifoong92.medium.com/beginners-guide-to-neural-speaker-diarization-with-pyannote-24ff4aa784b4).

```python
from subprocess import CalledProcessError, run
# I got an error the first time I ran it, but not the second...
from pyannote.audio import Pipeline
```

```python
def split_audio(input_file, output_file, start, end):
    length = end - start
    cmd = ["ffmpeg", "-ss", str(start), "-i", input_file, "-t", str(length), "-vn", "-acodec", "pcm_s16le", "-ar", "48000", "-ac", "1", output_file]
    try:
        run(cmd, capture_output=True, check=True).stdout
    except CalledProcessError as e:
        raise RuntimeError(f"FFMPEG error {str(e)}")
        
pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization@2.1",
    use_auth_token="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"
)
```

```python
for vid in videos['directors']:

    input_wav = "audio/" + vid + ".wav"
    output_dir = "pyannote_output"
    count = 10001

    # inference
    diarization = pipeline(input_wav)

    for turn, _, speaker in diarization.itertracks(yield_label=True):
        print(f"start={turn.start}s stop={turn.end}s speaker_{speaker}")

        speaker_dir = f"{output_dir}/speaker_{speaker}/"
        if not os.path.isdir(speaker_dir):
            os.mkdir(speaker_dir)

        filename = os.path.join(speaker_dir, f"interview-{count}.wav")
        split_audio(input_wav, filename, turn.start, turn.end)
        count += 1
```

Here I produced a list of files, ordered by when they occur:

```python
import os 
audio_dir = 'pyannote_output/comedians/'
audio_subfolders = os.listdir(audio_dir)

speaker_audio_dict = dict()

for subfolder in audio_subfolders:
    subfolder_path = os.path.join(audio_dir, subfolder)  # Create the full path to the subfolder
    if os.path.isdir(subfolder_path):  # Check if it's a directory
        subfolder_contents = os.listdir(subfolder_path)
        # print(subfolder_path)
        for item in subfolder_contents:
            # print(item)
            speaker_audio_dict.update({item: [subfolder_path.split('/')[-1], subfolder_path + '/' +  item]})

sorted_dict = sorted(speaker_audio_dict)[1:]
```

```python
from tqdm import tqdm
import speech_recognition as sr
import subprocess
import json
import os
import re
from pydub import AudioSegment
transcript = {}

def get_wav_duration(wav_file):
    # Load the WAV file using AudioSegment
    audio = AudioSegment.from_file(wav_file)

    # Get the duration in seconds
    duration_seconds = audio.duration_seconds

    return duration_seconds

def convert_wav_to_text(wav_file, time_end):
    # Run autosub and capture the output as a string
    output = subprocess.check_output(['autosub', wav_file, '--format', 'json'], stderr=subprocess.STDOUT, text=True)
    
    # Check if the output contains a reference to the JSON file
    json_file_match = re.search(r'Subtitles file created at (.+\.json)', output)
    if json_file_match:
        json_file_path = json_file_match.group(1)
        # Load the JSON file into a Python dictionary
        with open(json_file_path, 'r') as json_file:
            subtitle_dict = json.load(json_file)
        # Delete the JSON file
        os.remove(json_file_path)

        # Combine dictionary to a single recording:
        # subtitles = ' '.join(element['content'] for element in subtitle_dict)
        if subtitle_dict != []:
            seconds_elapsed_ = subtitle_dict[-1]['end']
            for dict_ in subtitle_dict:
                dict_.update({'start':dict_['start'] + time_end})
                dict_.update({'end':dict_['end'] + time_end})
        else:
            seconds_elapsed_ = get_wav_duration(wav_file)
                        
        return subtitle_dict, seconds_elapsed_

seconds_elapsed = 0
for interview in tqdm(sorted_dict):
    speaker = speaker_audio_dict[interview][0]
    filepath = speaker_audio_dict[interview][1]    

    text, seconds_elapsed_ = convert_wav_to_text(filepath, seconds_elapsed)
    seconds_elapsed += seconds_elapsed_
    transcript.update({interview:[speaker, text]})
    # print('{}: {}'.format(speaker, text))
    # print(text)
    
```

```python
# Specify the file path where you want to save the JSON file
file_path = "transcript.json"

# Write the dictionary to the JSON file
with open(file_path, "w") as json_file:
    json.dump(transcript, json_file, indent=4)

print(f"Transcripts have been saved to {file_path}")
```
